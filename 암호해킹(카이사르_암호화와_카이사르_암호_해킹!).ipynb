### 암호문 예외처리

- A~Z까지를 암호화 할 수 있는 데 암호화 키가 25라고 가정해봅시다.
- 이 경우 Z(25)와 Y(24)에 암호화 키를 더하면 그 결과는 50과 49가 됩니다.
- 그런데 우리가 암호화 할 수 있는 글자의 범위는 0부터 25 즉 A부터 Z입니다.
- 이런 경우는 어떻게 처리해줄까요?
- 구글 스프레드시트를 통해 연습해봅시다!

### 평문 예외처리

```python
# 카이사르 암호화
# https://www.nostarch.com/crackingcodes (BSD Licensed)

import pyperclip #모듈 불러오기

# 암호화/복호화 할 문장
message = 'This is my secret message.'
#message = 

# 암호화 키
key = 17

# 암호화(encrypt), 복호화(decrypt) 모드 설정
#mode = 'encrypt' # Set to either 'encrypt' or 'decrypt'.
mode = 'decrypt'

# 암호화가 가능한 모든 기호들:
SYMBOLS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890 !?.'

# 암호화/복화화 결과물을 저장
translated = ''

for symbol in message: #암호화/복호화 할 문장을 symbol이라는 이름으로 한 글자씩 돌면서 아래에 있는 것을 실행 
    # SYMBOLS에 있는 글자만 암호화 / 복호화 가능
    if symbol in SYMBOLS: #만약에 지금 확인하고 있는 글자가 암호화 / 복호화 가능하다면
        symbolIndex = SYMBOLS.find(symbol) #그 위치를 저장하기(숫자)

        # 암호화/복호화 모드 실행
        if mode == 'encrypt':
            translatedIndex = symbolIndex + key
        elif mode == 'decrypt':
            translatedIndex = symbolIndex - key

        # 원판 범위 벗어난 숫자와 음수 보정 (필요한 경우에만)
        if translatedIndex >= len(SYMBOLS): #만약에 키가 원판 범위보다 크다면
            translatedIndex = translatedIndex - len(SYMBOLS)
        elif translatedIndex < 0: #만약에 키가 원판 범위보다 작다면
            translatedIndex = translatedIndex + len(SYMBOLS)

        translated = translated + SYMBOLS[translatedIndex] #결과물에 한글자씩 암호화/복호호된 결과를 저장해줌
    else: 
        # 암호화/복호화 범위를 벗어난 글자라면 그냥 그대로 결과에 저장해줌
        translated = translated + symbol

# 결과물을 출력
print(translated)
#pyperclip.copy(translated)
```

### 카이사르 암호화 해킹 프로그램!!!

- 카이사르 암호화 암호는 어떻게 해킹할 수 있을까요?
- 해킹을 한다는 의미는 무엇일까요?
- 해킹을 한다는 것은 (암호문)을 (평문)로 만드는 것을 의미합니다.
- 그렇다면 어떻게 해킹을 해야 할까요?
- 무작위 대입법을 사용해봅시다!
- 키가 될 수 있는 모든 후보를 가지고 (암호문)를 (평문으)로 만들어보고
- 해석이 된다면 우리는 카이사르 암호를 해킹할 수 있을 것입니다


```python
# 카이사르 암호화 알고리즘 해킹 기법
# https://www.nostarch.com/crackingcodes (BSD Licensed)

message = 'guv6Jv6Jz!J6rp5r7Jzr66ntrM'
SYMBOLS = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890 !?.'
print(len(SYMBOLS)) #이것은 무슨 의미? (  원판의 개수          )

# 키가 될 수 있는 모든 숫자 범위를 하나씩 증가하면서 테스트한다
for key in range(len(SYMBOLS)): #암호화 키가 될 수 있는 범위(암호화 가능한 텍스트 길이)를 0부터 1씩 증가하면서 테스트 한다 
   # 하나의 후보를 처리할 때마다 후보 결과가 들어갈 translated를 공백으로 만든다
    translated = ''

    # 암호화된 messange의 글자 하나 하나를 symbol이라는 이름으로 처음부터 끝까지 순회한다:
    for symbol in message:
        if symbol in SYMBOLS: #만약에 해독이 가능하다면
            symbolIndex = SYMBOLS.find(symbol) #암호문의 위치에 해당하는 숫자를 저장하고 
            translatedIndex = symbolIndex - key #후보키만큼 뺴서 위치를 저장한다

            # 위치가 0보다 작을 때
            if translatedIndex < 0:
                translatedIndex = translatedIndex +  len(SYMBOLS)

            # 해독 결과물을 한글짜식 더해서 저장한다
            translated = translated + SYMBOLS[translatedIndex]

        else:
            # 해독이 불가능하다면 그대로 저장한다.
            translated = translated + symbol

    # 모든 후보 키 연산에 따른 해독 결과를 출력한다.
    print('Key #%s: %s' % (key, translated))
```

```python
#### 마찬가지로 위 프로그램을 개조해봅시다!
### 해독하려는 암호문을 직접 입력하고
### 암호화할 수 있는 글자 내용을 입력하면
### 이를 토대로 해킹을 통해 결과물을 출력하도록 바꿔봅시다!
```
